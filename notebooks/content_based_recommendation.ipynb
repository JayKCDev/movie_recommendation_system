## Content-based recommendation

```
import pandas as pd
movies = pd.read_csv("movies_small.csv", sep=";")
```

```
movies
```

```
# !pip install scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Find TF-IDF matrix using movies descriptions by text vectorization

```
tfidVectorizer = TfidfVectorizer(stop_words="english")
# movies["overview"] = movies["overview"].fillna("")
content = (
    movies["overview"].fillna("") + "" +
    (movies["genres"].fillna("") + "") * 3 +
    (movies["title"].fillna("") + "") * 2
)
print(type(content))
vectorized_matrix = tfidVectorizer.fit_transform(content)
# print(movies.shape)
print(len(movies.columns))
```

```
pd.DataFrame(vectorized_matrix.toarray(), columns = tfidVectorizer.get_feature_names_out())
```

### Calculate similarity matrix

```
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(vectorized_matrix, vectorized_matrix)
similarity_matrix
```

### FInd similar movies based on movie title

```
movie_title = "John Carter"
```

```
movie_index = movies.loc[movies["title"] == movie_title].index[0]
movie_index
```

```
scores = list(enumerate(similarity_matrix[movie_index]))
scores
```

```
def similar_movies(movie_title, numOfSimilars=3):
    matches = movies.loc[movies["title"] == movie_title]
    if matches.empty:
        raise ValueError(f"Movie '{movie_title}' not found in dataset")
    
    movie_index = matches.index[0]
    scores = list(enumerate(similarity_matrix[movie_index]))
    scores = sorted(scores, key=lambda x:x[1], reverse=True)
    similar_movie_indices = [scoreTup[0] for scoreTup in scores[1 : numOfSimilars + 1]]
    # similar_movies = movies.iloc[similar_movie_indices][["title", "release_date", "vote_average", "vote_count"]]
    # Alternative way below
    similar_movies = movies.loc[similar_movie_indices, ["title", "release_date", "vote_average", "vote_count"]]
    similar_movies_list = similar_movies.to_dict('records')
    return similar_movies_list
```

```
# movies["title"].head()
hah = similar_movies("Kung Fu Panda 3", 3)
hah
```
